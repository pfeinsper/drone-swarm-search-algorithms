## Learning curve considering a scenario with two agents

In the next two pictures we present the learning curve of reinforcement learning model and the size of each episode during the training phase. 

```python
df2 = pd.read_csv('./ray_results/DSSE_Coverage/PPO_2_agent_9/PPO_DSSE_Coverage_20516_00000_0_2025-02-21_00-23-51/progress.csv')
```

```python
media_movel_reward_acc = df2['env_runners/episode_reward_max'].rolling(window=100).mean()
lineplot(data=df2, x='timesteps_total', y=media_movel_reward_acc)
plt.xlabel('Timesteps')
plt.ylabel('Accumulated Reward')
plt.title('Learning curve for 2 agents')
plt.show()
```

```python
media_movel_tamanho_episodio = df2['env_runners/episode_len_mean'].rolling(window=100).mean()
lineplot(data=df2, x='timesteps_total', y=media_movel_tamanho_episodio)
plt.xlabel('Timesteps')
plt.ylabel('Episode length')
plt.title('Behavior of episode length for 2 agents')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

df4 = pd.read_csv('./ray_results/DSSE_Coverage/PPO_4_agent_9/PPO_DSSE_Coverage_5586c_00000_0_2025-02-21_08-39-16/progress.csv')

media_movel_reward_acc = df2['env_runners/episode_reward_max'].rolling(window=50).mean()
media_movel_tamanho_episodio = df2['env_runners/episode_len_mean'].rolling(window=50).mean()

media_movel_reward_acc4 = df4['env_runners/episode_reward_max'].rolling(window=50).mean()
media_movel_tamanho_episodio4 = df4['env_runners/episode_len_mean'].rolling(window=50).mean()


# Plotar a média móvel da recompensa acumulada
fig, ax1 = plt.subplots(figsize=(10, 6))
line1 = sns.lineplot(data=media_movel_reward_acc, color="g", ax=ax1)
line3 = sns.lineplot(data=media_movel_reward_acc4, color="r", ax=ax1)
ax1.set_xlabel('Episodes')
ax1.set_ylabel('Accumulated Reward')
#ax1.tick_params(axis='y', labelcolor='g')

# Criar um segundo eixo y para o tamanho do episódio
ax2 = ax1.twinx()
line2 = sns.lineplot(data=media_movel_tamanho_episodio, color="b", ax=ax2)
line4 = sns.lineplot(data=media_movel_tamanho_episodio4, color="y", ax=ax2)
ax2.set_ylabel('Episode Length')
#ax2.tick_params(axis='y', labelcolor='b')

# Adicionar título
#plt.title('Learning curve and episode length for 2 and 4 agents')

# Adicionar legenda

legend_handles = [
    Line2D([0], [0], color='g', lw=2, label='Accumulated Reward 2 agents'),
    Line2D([0], [0], color='b', lw=2, label='Episode Length 2 agents'),
    Line2D([0], [0], color='r', lw=2, label='Accumulated Reward 4 agents'),
    Line2D([0], [0], color='y', lw=2, label='Episode Length 4 agents')
]

# Adicionar legenda
ax1.legend(handles=legend_handles, loc='center right')

#lines = [line1.get_lines()[0], line2.get_lines()[0], line3.get_lines()[0], line4.get_lines()[0]]
#labels = ['Accumulated Reward 2 agents', 'Episode Length 2 agents', 'Accumulated Reward 4 agents', 'Episode Length 4 agents']
#ax1.legend(lines, labels, loc='upper right')

# Mostrar o gráfico
plt.show()
```

## Model's behavior in a scenario with two agents

In the next pictures we will present the behavior of the model and other algortihms in a scenario with two agents.

```python
a_start_2 = pd.read_csv('results/a_star_2.csv')
expanding_2 = pd.read_csv('results/expanding_search_2_agent.csv')
traditional_2 = pd.read_csv('results/traditional_search_2_agent.csv')

a_start_2['Algol'] = 'A*'
expanding_2['Algol'] = 'Expanding Square Search'
traditional_2['Algol'] = 'Parallel Line Scanning'

combined_df = pd.concat([a_start_2, expanding_2, traditional_2], ignore_index=True)
combined_df
```

```python

lineplot(data=combined_df, x='step', y='accumulated_pos', hue='Algol')
plt.title('The Accumulated POS for an environment with 2 agents')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative POS')
plt.legend(title='Algorithms')
plt.show()
```

```python
lineplot(data=combined_df, x='step', y='coverage_rate', hue='Algol')
plt.title('The Coverage Rate for an environment with 2 agents')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative Coverage Rate')
plt.legend(title='Algorithm')
plt.show()
```

```python
lineplot(data=combined_df, x='step', y='repeated_coverage', hue='Algol')
plt.title('The Repeated Coverage for an environment with 2 agents')
plt.xlabel('Timesteps')
plt.ylabel('Repeated Coverage')
plt.legend(title='Algorithm')
plt.show()
```
