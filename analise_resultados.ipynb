## Model's behavior in a scenario with two agents

In the next pictures we will present the behavior of the model and other algortihms in a scenario with two agents.

```python
a_start_2 = pd.read_csv('results/a_star_2.csv')
expanding_2 = pd.read_csv('results/expanding_search_2_agent.csv')
traditional_2 = pd.read_csv('results/traditional_search_2_agent.csv')

a_start_2['Algol'] = 'A*'
expanding_2['Algol'] = 'Expanding Square Search'
traditional_2['Algol'] = 'Parallel Line Scanning'

combined_df = pd.concat([a_start_2, expanding_2, traditional_2], ignore_index=True)
combined_df
```

```python

lineplot(data=combined_df, x='step', y='accumulated_pos', hue='Algol')
plt.title('The Accumulated POS for an environment with 2 agents')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative POS')
plt.legend(title='Algorithms')
plt.show()
```

```python
lineplot(data=combined_df, x='step', y='coverage_rate', hue='Algol')
plt.title('The Coverage Rate for an environment with 2 agents')
plt.xlabel('Timesteps')
plt.ylabel('Cumulative Coverage Rate')
plt.legend(title='Algorithm')
plt.show()
```

```python
lineplot(data=combined_df, x='step', y='repeated_coverage', hue='Algol')
plt.title('The Repeated Coverage for an environment with 2 agents')
plt.xlabel('Timesteps')
plt.ylabel('Repeated Coverage')
plt.legend(title='Algorithm')
plt.show()
```
